```python
import os
import pandas as pd

directory = os.getcwd()

def read_json_files_to_dataframe(container_path):
    # Extracting the JSON file names
    json_files = [file for file in os.listdir(container_path) if file.endswith(".json")]

    # Loading the JSON files into a list of DataFrames
    dfs = []
    for json_file in json_files:
        file_path = os.path.join(container_path, json_file)
        with open(file_path, "r", encoding="utf-8") as f:
            df = pd.read_json(f, lines=True)
            dfs.append(df)

    # Combining the DataFrames
    combined_df = pd.concat(dfs)

    return combined_df

def save_dataframe_to_json(df, output_path):
    df.to_json(output_path, orient="records", lines=True, force_ascii=False)

if __name__ == "__main__":
    container_path = directory + "/raw/"

    # Read JSON files to DataFrame
    combined_df = read_json_files_to_dataframe(container_path)

    # Convert combined DataFrame to a JSON file
    output_json_file = os.path.join(container_path, "combined_data.json")
    save_dataframe_to_json(combined_df, output_json_file)

    print("Conversion completed successfully. JSON file saved at:", output_json_file)
```
